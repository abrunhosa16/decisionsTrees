```python
import pandas as pd, numpy as np, sklearn, matplotlib as plt

class TreeNode:
    def __init__(self, condition=None, value=None, children=[]) -> None:
        self.value = value
        self.children = children
        self.condition = condition
        
    def is_leaf(self) -> bool:
        return len(self.children) == 0
    
    def add_children(self, children=[]) -> None:
        self.children += children

def entropy(df: pd.DataFrame, attribute: str) -> float:
    if len(df) == 0 or len(df[attribute].unique()) == 1:
        return 0 

    # Implementação vetorizada para aumentar eficiencia 
    # normalize = true faz a proporção dos valores 
    value_counts = df[attribute].value_counts(normalize=True)
    entropy = -(value_counts * np.log2(value_counts)).sum()

    return entropy

def conditional_entropy(df:pd.DataFrame, attribute:str, target_attribute:str) -> float: #H( attribute | target_attribute )
    target_classes = df[target_attribute].unique()
    entropy_attribute = 0
    total_examples = len(df)

    for cls in target_classes:
        subset_df = df[df[target_attribute] == cls]
        entropy_subset = entropy(subset_df, attribute)
        prob_cls = len(subset_df) / total_examples #probabilidade desta classe acontecer P(cls)
        entropy_attribute += prob_cls * entropy_subset
    return entropy_attribute

def information_gain(df:pd.DataFrame, attribute:str) -> float:
    return entropy(df, 'Class') + conditional_entropy(df, 'Class', attribute)
```

```python
import math
import numpy as np 
import random   

def yesNo(df, target:str):
        a = df[target].value_counts()[0] / len(df)
        return a, 1-a

def subsets(df, target):
        yes = df[df[target] == 'yes']
        no = df[df[target] == 'no']
        return yes, no

def split_representative(df, target: str, perc: float) -> (list, list, list, list):
        yes_per, no_per = yesNo(df, target) # proporção de sims e naos
        yes_sub, no_sub = subsets(df, target)[0].index.values.tolist(),  subsets(df, target)[1].index.values.tolist()# o subconjunto de sins e naos representados pelo ID
        # Embaralhar os indices de sins e naos
        total_yes, total_no = np.random.choice(yes_sub, size=len(yes_sub), replace=False), np.random.choice(no_sub, size=len(no_sub), replace=False)
        num_yes, num_no = math.ceil(perc * yes_per * len(yes_sub)), math.ceil(perc * no_per* len(no_sub)) # calcula o valor a ser removido para teste 
        test = sorted([*total_yes[:num_yes],*total_no[:num_no]]) # indices para teste
        train = sorted([*total_yes[num_yes:], *total_no[num_no:]]) # indices para treino 

        test_x = df.iloc[test]
        train_x = df.iloc[train]
        return test_x, train_x



#split_representative(df, 'Play', 0.2)[0]
split_representative(df, 'Play', 0.2)[1]

```
